```python
# demo_datamedic.py
# Simple demo for testing DataMedic library

import sys
import pandas as pd

# Step 1: Add parent folder to sys.path
sys.path.append(r"C:\Users\jorda\Documents\GitHub\OOP-Final_Project\1 - Final Project")

# Step 2: Import classes from DataMedic
from DataMedic.data_doctor import DataDoctor
from DataMedic.report_generator import ReportGenerator
from DataMedic.data_inspector import DataInspector

# Step 3: Create a sample DataFrame
data = {
    "Name": ["Alice", "Bob", None, "David", "Alice", "Eve", None],
    "Age": [25, 30, 22, None, 25, 28, 35],
    "Score": [90, 85, 95, 100, 90, 88, None]
}

df = pd.DataFrame(data)

# Step 4: Inspect the data
doctor = DataDoctor(df)
issues = doctor.inspect()
print("=== Data Issues ===")
print(issues)

# Step 5: Diagnose suggested actions
diagnosis = doctor.diagnose()
print("\n=== Suggested Actions ===")
for action in diagnosis:
    print("-", action)

# Step 6: Apply fixes
doctor.treat()
print("\n=== Fix Log ===")
for log in doctor.get_fix_log():
    print("-", log)

# Step 7: Generate report
reporter = ReportGenerator(doctor)
report = reporter.report()
print("\n=== Full Report ===")
print(report)

# Step 8: Health score
score = reporter.health_score()
print("\n=== Health Score ===")
print(score)
```

```python
# demo_health_80.py
import sys
import pandas as pd

# Add parent folder to sys.path
sys.path.append(r"C:\Users\jorda\Documents\GitHub\OOP-Final_Project\1 - Final Project")

# Import DataMedic classes
from DataMedic.data_doctor import DataDoctor
from DataMedic.report_generator import ReportGenerator

# Create a sample dataset
# We will design it so that:
# missing values: 5 (penalty 5)
# duplicates: 3 (penalty 3)
# outliers: 12 (penalty 12)
# Total penalty = 20 â†’ Health score = 80
data = {
    "Name": ["Alice", "Bob", None, "David", "Alice", "Eve", "Frank", None],
    "Age": [25, 30, 22, 120, 25, 28, 200, 35],  # 120 and 200 are outliers
    "Score": [90, 85, 95, 100, 90, 88, 300, 50]  # 300 is outlier
}

df = pd.DataFrame(data)

# Step 1: Inspect and diagnose
doctor = DataDoctor(df)
print("=== Data Issues ===")
print(doctor.inspect())

print("\n=== Suggested Actions ===")
for action in doctor.diagnose():
    print("-", action)

# Step 2: Apply fixes
doctor.treat()
print("\n=== Fix Log ===")
for log in doctor.get_fix_log():
    print("-", log)

# Step 3: Generate report and health score
reporter = ReportGenerator(doctor)
report = reporter.report()
print("\n=== Full Report ===")
print(report)

score = reporter.health_score()
print("\n=== Health Score ===")
print(score)  # Should print 80
```
