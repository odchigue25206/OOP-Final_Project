```python
import pandas as pd
from datamedic import DataDoctor, ReportGenerator

df = pd.DataFrame({
    'Name': ['A', 'B', None],
    'Age': [20, None, 30],
    'Score': [100, 200, 5000]  # naay outlier
})

doctor = DataDoctor(df)
doctor.inspect()
doctor.diagnose()
doctor.treat()

report = ReportGenerator(doctor)
print(report.report())
print("Health Score:", report.health_score())
```
