```python
#updated_main.ipynb
import os
print(os.getcwd())
print(os.listdir())

# Removed sys.path.append since DataMedic is now a proper package
from DataMedic import DataDoctor, ReportGenerator, DataInspector, DataGetter  # Updated imports
import pandas as pd
from io import StringIO

def get_sample_df():
    """Returns sample DataFrame to avoid inline redundancy."""
    return pd.DataFrame({
        "Name": ["Alice", "Bob", None, "David", "Alice"],
        "Age": [25, 30, 22, None, 25],
        "Score": [90, 85, 95, 100, 90]
    })

def print_summary(doctor, reporter):
    """Prints a structured summary of data processing."""
    print("Inspection:", doctor.inspect())
    print("Diagnosis:", doctor.diagnose())
    doctor.treat()
    print("Fix Log:", doctor.get_fix_log())
    report = reporter.report()
    print("Report:", report)
    print("Health Score:", reporter.health_score())

# Load DataFrame
getter = DataGetter()
file_name = "data.csv"
content = getter.read_file(file_name)
df = pd.read_csv(StringIO(content)) if isinstance(content, str) and not content.startswith("Error") else (print(content) or get_sample_df())

# Process and report
doctor = DataDoctor(df)
reporter = ReportGenerator(doctor)
print_summary(doctor, reporter)
```
